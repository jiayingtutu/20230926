```python
import random
ans = random.sample(range(1,10),4)
print(ans)
a = b = n = 0
while a!=4:
  a = b = n = 0
  guess = list(input("輸入四個數字(不會重複)"))
  if len(guess) > 4:
    guess = list(input("輸入數字過多，請輸入四個數字(不會重複)"))
  else:
    guess = guess
  for i in guess:
    if int(guess[n]) == ans[n]:
      a += 1
    else:
      if int(i) in ans:
        b += 1
    n += 1
  output = ','.join(guess).replace(',','')
  print(f'{output}: {a}A{b}b')
print("BINGO!")
```
