```python
a=int(input('請輸入大於1的正整數'))
if(a<=1):
  a = int(input("請重新輸入一個大於一的正整數"))
x=0
for b in range(1,(a+1)):
   n=0
   for i in range(1,(b+1)):
      if b%i==0:
        n+=1
   if n==2:
     x=b
print(x)
```

```python
n=input('請輸入數字:')
n=n.split(',')
L=len(n)
b=[]
for i in range(0,L):
  a=0
  for j in range(0,L):
    if int(n[i])>int(n[j]):
      a+=1
  b.append(a)
print(b)
```

```python
a = input("請輸入一串由零與一組成的數列")
b = len(a)
ans = []
for i in range (0,b):
  num = 0
  for j in range (0,b):
    if (int(a[j])==1):
      if(j>i):
        num+=j-i
      else:
        num+=i-j
  ans.append(num)
print(ans)
```
